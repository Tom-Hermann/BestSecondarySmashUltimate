```python
def getBestSecondarys(id):
    bestSecond = dict(zip(df.columns.values[1:], [0]*len(df.columns.values[1:])))
    for i, muValue in enumerate(df.iloc[id]):
        try:
            muValue = float(muValue)
            if muValue < 0:
                # print(value, df.columns.values[i], df.iloc[i -1].to_string(index=False))
                for j, muMUValue in enumerate(df.iloc[i - 1]):
                    try:
                        muMUValue = float(muMUValue)
                        if muMUValue < 0:
                            # print(muMUValue, df.columns.values[j])
                            bestSecond[df.columns.values[j]] += muMUValue * muValue
                    except:
                        pass
        except:
            pass

    bestSecond = {x:round(y, 2) for x,y in bestSecond.items() if y!=0}
    bestSecond = sorted(bestSecond.items(), key=lambda x: x[1], reverse=True)
    return bestSecond
```

```python
def getNegativeMU(id):
    negatifMU = []
    for i, muValue in enumerate(df.iloc[id]):
        try:
            muValue = float(muValue)
            if muValue < 0:
                negatifMU.append(i)
        except:
            pass
    return negatifMU

```

```python

def expandBestSecondaryResult(id):
    bestSecond = getBestSecondarys(id)
    negatifMU = getNegativeMU(id)

    print("Worst Matchup: ", *[df.columns.values[i] for i in negatifMU], sep=", ")

    for i in bestSecond:
        print(f"{i[0]}: {i[1]}")
        persoId = (df.columns.get_loc(i[0])) - 1
        for j in negatifMU:
            print(f"\t vs {df.columns.values[j]}: {df.iloc[persoId][j]}")
```

```python
expandBestSecondaryResult(0)
```
